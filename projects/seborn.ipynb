```python
#The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

#Import the stata library from numpy
from scipy import stats

#These are the plotting modulees adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#command so that plots appear in the jupyter notebook
%matplotlib inline
```

```python
#create a random normal-dist dataset
dataset1 =randn(100)

#Plot a histogram of the dataset, note bins=10 by default
plt.hist(dataset1)
```

```python
#Let's make another dataset
dataset2 =randn(80)
#Plot
plt.hist(dataset2,color='indianred')
```

```python
#We can use normed to plot on same plot

#set normed=True for the plots to be normalized in order to compare datasets with differnt numbers of observations
#set alpha=0.5 for transparency

plt.hist(dataset1,normed=True,color='indianred', alpha=0.5,bins=20)
plt.hist(dataset2,normed=True,alpha=0.5,bins=20)
```

```python
#make 2 more random normal dist data sets
data1 = randn(1000)
data2 = randn(1000)

sns.jointplot(data1,data2)
```
